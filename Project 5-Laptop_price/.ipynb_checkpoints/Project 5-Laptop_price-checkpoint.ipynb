## Business Case
To create a model which can estimate the price of laptops based on their attributes.

## Domain Analysis
- laptop_ID : ID of each laptop
- Company : Name of the company which manufactured the laptops
- Product : Model name of the laptops
- TypeName : To which category a laptop belongs to
- Inches : Size of screen 
- ScreenResolution : Type of screen and pixel dimension
- Cpu : Central Processing Unit
- Ram : Main memory in each laptop
- Memory : Storage additional to RAM in SSD (Solid State Drive) or HDD (Hard Disk Drive)
- Gpu : Graphics Processing Unit
- OpSys : Name of operating system of laptops
- Weight : How much each laptop weights
- Price_in_euros : Price of each laptop in Euros

## Data Preprocessing

### Checking for Missing Values

```python
df.isnull().sum()
```

There is no any missing value present in any column.

```python
df["Ram"].value_counts()
```

```python
df["Ram"]=df["Ram"].str.replace("GB","")
```

```python
df["Ram"]=df["Ram"].astype("int32")
```

```python
df.info()
```

```python
df["Weight"].value_counts()
```

```python
df["Weight"]=df["Weight"].str.replace("kg","")   # to remove kg and only keep digit part 
```

```python
df["Weight"]=df["Weight"].astype("float")
```

```python
df["ScreenResolution"].value_counts()
```

Seperating x and y resolution so that we can create a new column named PPI. PPI stands for pixel per inches.

```python
new_screen=df["ScreenResolution"].str.split("x",n=1,expand=True)
new_screen  # n: Numbers of max separations to make in a single string, default is -1 which means all.
    # expand: Boolean value, returns a data frame with different value in different columns if True.
    # Else it returns a series with list of strings. 
```

```python
df['x_resolution']=new_screen[0]
df['x_resolution']=df['x_resolution'].str.strip()

df['y_resolution']=new_screen[1]
```

```python
df.head()
```

```python
df['x_resolution']=df['x_resolution'].str.replace(r"\D","") # replacing the text part of column x_resolution to get only digit part
```

```python
df.head()
```

```python
df.info()
```

```python
df['x_resolution']=df['x_resolution'].astype("int32")
```

```python
df['y_resolution']=df['y_resolution'].astype("int32")
```

```python
df['PPI']= ((df['x_resolution']**2)+(df['y_resolution']**2))**0.5/df['Inches']
```

```python
df.drop(["ScreenResolution","x_resolution","y_resolution","Inches"],axis=1,inplace=True)   # dropping ScreenResolution, x_resolution, y_resolution and Inches column as they are no longer needed
```

```python
df.info()
```

```python
df["Memory"].value_counts()
```

```python
df["Memory"]=df["Memory"].str.replace("GB","")
```

```python
df["Memory"]=df["Memory"].str.replace("TB","000")
```

```python
df["Memory"]
```

```python
new_storage=df["Memory"].str.split("+",n=1,expand=True)
pd.set_option('display.max_rows', None)
new_storage
```

```python
df['storage1']=new_storage[0]
df['storage1']=df['storage1'].str.strip()               # strip(): it removes whitespaces if any

df['storage2']=new_storage[1]
```

```python
df.head()
```

```python
# df.corr()['Price_in_euros']
```

```python
# can see there are 4 categories in storage 1. so we will make 4 columns

df["SSD1"]=df["storage1"].apply(lambda x:1 if "SSD"in x else 0)
df["HDD1"]=df["storage1"].apply(lambda x:1 if "HDD"in x else 0)
df["Hybrid1"]=df["storage1"].apply(lambda x:1 if "Hybrid"in x else 0)
df["Flash Storage1"]=df["storage1"].apply(lambda x:1 if "Flash Storage"in x else 0)
```

```python
df.head(2)
```

```python
df['storage1']=df['storage1'].str.replace(r"\D","")
```

```python
df.head(2)
```

```python
df['storage2'].isnull().sum()
```

```python
df['storage2'].isnull().sum()
```

```python
df["SSD2"]=df["storage2"].apply(lambda x:1 if "SSD"in x else 0)
df["HDD2"]=df["storage2"].apply(lambda x:1 if "HDD"in x else 0)
df["Hybrid2"]=df["storage2"].apply(lambda x:1 if "Hybrid"in x else 0)
df["Flash Storage2"]=df["storage2"].apply(lambda x:1 if "Flash Storage"in x else 0)
```

```python
df.head(2)
```

```python
# Combining Flash Storage into a single column
df["Flash Storage"]=(df['storage1']*df["Flash Storage1"] + df["storage2"]*df["Flash Storage2"])
```

### Hyperparameter tuning of Decision Tree

```python
from sklearn.model_selection import GridSearchCV

params = {
    "criterion":("gini", "friedman_mse", "squared_error", "absolute_error"), #quality of split
    "splitter":("best", "random"), # searches the features for a split
    "max_depth":(list(range(1, 40))), #depth of tree range from 1 to 19
    "min_samples_split":[2, 3, 4, 6,8, 10, 12],    #the minimum number of samples required to split internal node
    "min_samples_leaf":list(range(1, 20)),#minimum number of samples required to be at a leaf node,we are passing list which is range from 1 to 19 
    "max_features": ("auto","sqrt","log2")
}


tree_clf = DecisionTreeRegressor(random_state=3)#object creation for decision tree with random state 3
tree_cv = GridSearchCV(tree_clf, params, scoring="r2", n_jobs=-1, verbose=1, cv=4)




#passing model to gridsearchCV ,
#tree_clf-->model
#params---->hyperparametes(dictionary we created)
#scoring--->performance matrix to check performance
#n_jobs---->Number of jobs to run in parallel,-1 means using all processors.
#verbose=Controls the verbosity: the higher, the more messages.
#>1 : the computation time for each fold and parameter candidate is displayed;
#>2 : the score is also displayed;
#>3 : the fold and candidate parameter indexes are also displayed together with the starting time of the computation.
#cv------> number of flods




tree_cv.fit(x_train,y_train)#training data on gridsearch cv
best_params = tree_cv.best_params_#it will give you best parameters 
print(f"Best paramters: {best_params})")#printing  best parameters

```

### Hyperparameter tuning of Random Forest

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]#List Comprehension-using for loop in list
criterion = ["squared_error", "absolute_error", "friedman_mse", "poisson"]

max_features = ['log2', 'sqrt']#maximum number of features allowed to try in individual tree
max_depth = [int(x) for x in np.linspace(10, 110, num=11)]#List Comprehension-using for loop in list
max_depth.append(None)
min_samples_split = [2, 5, 10, 12, 14, 18, 22]#minimum number of samples required to split an internal node
min_samples_leaf = [1, 2, 4, 6, 8, 9, 10, 12]#minimum number of samples required to be at a leaf node.
bootstrap = [True, False]#sampling 

#dictionary for hyperparameters
random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}

rf_clf1 = RandomForestRegressor(random_state=42)#model

rf_cv = RandomizedSearchCV(estimator=rf_clf1, scoring='r2',param_distributions=random_grid, n_iter=100, cv=5, 
                               verbose=2, random_state=42, n_jobs=-1) 
#estimator--number of decision tree
#scoring--->performance matrix to check performance
#param_distribution-->hyperparametes(dictionary we created)
#n_iter--->Number of parameter settings that are sampled. n_iter trades off runtime vs quality of the solution.default=10
##cv------> number of flods
#verbose=Controls the verbosity: the higher, the more messages.
#n_jobs---->Number of jobs to run in parallel,-1 means using all processors.

rf_cv.fit(x_train, y_train)##training data on randomsearch cv
rf_best_params = rf_cv.best_params_##it will give you best parameters 
print(f"Best paramters: {rf_best_params})")##printing  best parameters
 
```
